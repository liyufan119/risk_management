```python
import glob

path = r'E:\workplace\py36\风控_2\data\一级分类\0315' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, usecols=[1,2,3,4,5,6,8,9],converters = {'a.store_id': str,'a.user_id': str, 'a.spu_sn': str } )
    li.append(df)

category_0315 = pd.concat(li, axis=0, ignore_index=True)
```
