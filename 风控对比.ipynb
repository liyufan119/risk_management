```python
import pandas as pd
import numpy as np
```

```python
store_list = pd.read_excel('风控名单未标记的门店.xlsx',converters = {'门店ID': str})
```

```python
store_list.dtypes
```

```python
store_list.head()
```

```python
store_list['customers_diff'] = store_list['下单人数']-store_list['下单微信数']
```

```python
store_list.describe(percentiles=[.75,.80,.85,.9,.95])
```

```python
#切换账号大于18的门店
store_list_gt18=store_list.loc[store_list.customers_diff>18].sort_values(by=['总金额','下单人数'], axis=0, ascending=[False,True])
```

```python
store_list_gt18.to_excel('切换账号大于18个的门店.xlsx')
```
